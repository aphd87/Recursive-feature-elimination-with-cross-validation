```python
#Recursive feature elimination with cross validation
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LinearRegression
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
#Fits model that starts with all input variables, and works to create a strong model
#Algorithm can determine which determined the best levels of accuracy
my_df = pd.read_csv("feature_selection_sample_data.csv")
```

```python
#Split data into X and y objects

X = my_df.drop(["output"], axis=1) #axis=1 drop column
y = my_df["output"]
```

```python
regressor = LinearRegression()
feature_selector = RFECV(regressor)
```

```python
#By default, cross validation uses 5 chunks of data to analyze
#Cross validation  - split into 5, training on model on 4, and validating model on remaining chunk
#Powerful way to validate model

fit = feature_selector.fit(X,y)
```

```python
optimal_feature_count = feature_selector.n_features_
```

```python
print(f"Optimal number of features: {optimal_feature_count}")
```

```python
X_new = X.loc[:, feature_selector.get_support()]
```

```python
X_new
#2 optimal variables - input1, input2
#Through cross validation - input3 and input4 were not useful here
```

```python
plt.plot(range(1, len(fit.cv_results_['mean_test_score']) +1), fit.cv_results_['mean_test_score'], marker="o")
#range = X values to go from 1 feature to 4, rather than 0 to 3
#fit.grid_scores_ = Contains accuracy scores for each input
#marker = o - aesthetically pleasing plot
plt.ylabel("Model score")
plt.xlabel("Number of Features")
plt.title(f"Feature Selection suing RFE \n Optimal number of features is {optimal_feature_count} (at score of {round(max(fit.cv_results_),4)})")
```

```python
plt.plot(range(1, len(fit.cv_results_['mean_test_score']) + 1), fit.cv_results_['mean_test_score'], marker="o")
plt.ylabel("Cross-validation score")
plt.xlabel("Number of Features")
plt.title(f"Feature Selection using RFE\nOptimal number of features is {optimal_feature_count} (at score of {round(max(fit.cv_results_['mean_test_score']), 4)})")
plt.tight_layout()
plt.show()
```
